```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Read uploaded CSV file
import io
df = pd.read_csv(io.BytesIO(uploaded['material.csv']))
```

```python
# Basic Info
print("Dataset Summary:")
print(df.describe(include='all'))
```

```python
# Set plot style
sns.set(style="whitegrid")
```

```python
# Histogram of Strengths
plt.figure(figsize=(12, 5))
sns.histplot(df['Su'], bins=30, kde=True, color='skyblue')
plt.title('Ultimate Tensile Strength (Su) Distribution')
plt.xlabel('Su (MPa)')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Su', y='Sy', hue='Use', data=df)
plt.title('Yield Strength vs Ultimate Strength')
plt.xlabel('Su (MPa)')
plt.ylabel('Sy (MPa)')
plt.legend(title='Use')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 6))
corr = df.drop(columns=['Material', 'Use']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Between Material Properties')
plt.show()
```

```python

# Density vs Young's Modulus
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Ro', y='E', hue='Use')
plt.title("Density vs Young's Modulus")
plt.xlabel("Density (kg/m)")
plt.ylabel("Young's Modulus (MPa)")
plt.show()
```
